# Exploratory Analysis

## Age Distribution 

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
# The above charts shows that people of 25 were more in ship while people above age 70 were least.
```

## No of passangers according to Sex

```python
sns.countplot(x='Sex', data=df)
plt.title('Count of Passengers by Sex')
plt.show()
```

```python
# Analysis shows that there were more male passengers than female passengers
```

## Survival V/S Gender

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

```python
#Analysis shows that there were more males survived in tragic incident compared to females 
```

## No. of people survived wrt to passenger class

```python
sns.countplot(x='Survived', hue='Passanger class', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
#Survived:
#Among persons who survived, most were from passanger class 3 while least were from passanger class 1
# Dead:
#among those who died most of them were from passenger class 3 almost 70 and least were from passanger class 2 almost 35
#It shows that most people were from passanger class 3
```

## No of passanger survived with various conditions

```python
sns.pairplot(df, hue='Survived')
plt.show()
```

## Embarked Place Analysis

```python
result = df.groupby(['Embarked', 'Sex']).agg(
    Total_Embarked = ('Sex', 'count'),
    Survived_Count = ('Survived', 'sum')
).reset_index()

print(result)
```
